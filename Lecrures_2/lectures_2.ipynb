```python
import numpy as np
import matplotlib.pylab as plt
```

# Устранимый разрыв  $$ y =\frac{x^2 - 4}{x-2} $$

```python
def func1(arg):
    return (arg**2 - 4)/(arg - 2)
x = np.linspace(-2,3)
y = func1(x)
epsilon = 0.05
mask = np.abs(x-2)<epsilon
plt.plot(x,y, 'black')
plt.scatter(x[mask], y[mask], color='orange', s=500, marker='o')
plt.xlabel('x')
plt.ylabel('y')
```

# Разрыв 1го рода  $$ y =\frac{|x - 2|}{x-2} $$

```python
def func2(arg):
    return np.abs(arg-2)/(x-2)
x = np.linspace(-2,10)
y = func2(x)
epsilon = 0.3
mask = np.abs(x-2)<epsilon
plt.plot(x,y, 'o', 'black')
plt.scatter(x[mask], y[mask], color='orange', s=700, marker='o')
```

# Разрыв второго рода  $$ y =\frac{1}{(x-2)^2} $$

```python
def func3(arg):
    return (arg-2)**(-2)
x = np.linspace(-2,5)
y = func3(x)
epsilon = 0.05
mask = x == 2
plt.plot(x,y, 'black')
plt.scatter(x[mask], y[mask], color='orange', s=700, marker='o')
```

# Предел функции $$ y = x^2$$

```python
def func4(arg):
    return arg**2
x = np.linspace(2,4)
y = func4(x)
Point = 3
mask = x == Point
epsilon = 1
delta = np.sqrt(epsilon + func4(Point)) - Point
plt.axvline(x=Point - delta, color = 'g')
plt.axvline(x=Point + delta, color = 'g')
plt.axhline(y = func4(Point) - epsilon, color = 'b')
plt.axhline(y = func4(Point) + epsilon, color = 'b')
plt.scatter(Point,func4(Point), color='red', s=100, marker='o')
plt.plot(x,y, 'black')

```

# Сумма пределов $$ \lim_{x \to a} \big( \frac{x^2 - 4}{x-2} + \frac{(x^2 - 4)^2}{x-2} \big)$$

```python
def func5(arg):
    return  (arg**2 - 4)/(arg - 2) + ((arg**2 - 4)**2)/(arg - 2)
x = np.linspace(-2,3)
y = func5(x)
epsilon = 0.05
mask = np.abs(x-2)<epsilon
plt.plot(x,y, 'black')
plt.scatter(x[mask], y[mask], color='red', s=100, marker='o')
```

# Разность пределов $$ \lim_{x \to a} \big( \frac{x^2 - 4}{x-2} - \frac{(x^2 - 4)^2}{x-2} \big)$$

```python
def func6(arg):
    return (arg**2 - 4)/(arg - 2) - ((arg**2 - 4)**2)/(arg - 2)
x = np.linspace(-2,3)
y = func6(x)
epsilon = 0.05
mask = np.abs(x-2)<epsilon
plt.plot(x,y, 'black')
plt.scatter(x[mask], y[mask], color='red', s=100, marker='o')
```

# Бесконечно малые функции $$ f(x) = \frac{1}{x}$$

```python
n = 120
epsilon = 0.01
chastota = n
nomer_N = []
x = np.linspace(1, n,chastota)
y = 1  / x 

x1 = np.linspace(1, n,chastota)

y2 = [ + epsilon]*chastota
y3 = [0.0]*chastota

redX = []
redY = []

for i in range(len(x)):
    if abs(y[i]-y3[i]) <= epsilon:
        nomer_N.append(i)
        redX.append(x[i])
        redY.append(y[i])
print(nomer_N)
plt.plot(x, y, 'o', color='black')
plt.plot(redX, redY, 'o', color='red')
plt.plot(x1, y2, color='g', label='epsilon')
plt.plot(x1, y3, color='b', label='epsilon')
plt.xlabel("n")
plt.ylabel("")
plt.show()
```

# Бесконечно малые разного порядка $$ 1-x, (1-x)^2 $$

```python
rcParams['figure.figsize'] = [11,11]
def func_circle(arg):
    return 1-arg
def line_function(arg):
    return (arg-1)**2
x = np.linspace(0.,1.1)
y1 = func_circle(x)
y2 = line_function(x)
epsilon = 0.05
mask = np.abs(x-2)<epsilon
plt.plot(x,y1, 'red')
plt.plot(x,y2, 'green')
```

# Бесконечно малые функции разного порядка $$ (x - 1)^4$$ $$ (x - 1)^6$$ $$ (x - 1)^{100}$$

```python
rcParams['figure.figsize'] = [11,11]
def func_circle(arg):
    return (arg-1)**4*np.sin(arg)
def line_function(arg):
    return (arg - 1)**6
def function3(arg):
    return (arg - 1)**8
x = np.linspace(0.5,1.2)
y1 = func_circle(x)
y2 = line_function(x)
y3 = function3(x)
epsilon = 0.05
mask = np.abs(x-2)<epsilon
plt.plot(x,y1, 'red')
plt.plot(x,y2, 'green')
plt.plot(x,y3, 'blue')
```
